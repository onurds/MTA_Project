```python

import json
import re
import random
from collections import Counter, defaultdict
from typing import Dict, List, Tuple, Optional

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.metrics import classification_report, precision_recall_fscore_support
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence
from tqdm.notebook import tqdm
import os

# Set seeds for reproducibility
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)

print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
```


## 1. Data Loading and Preprocessing
Load the silver relations dataset, parse span and relation columns, and inspect a small sample.


```python

# Paths
silver_path = "Preprocessed/MTA_Data_silver_relations.csv"

# Load dataset
silver_df = pd.read_csv(silver_path)

# Utility to parse JSON-like columns

def parse_json_column(series: pd.Series) -> pd.Series:
    def _parse(x):
        if pd.isna(x):
            return []
        if isinstance(x, list):
            return x
        x = str(x).strip()
        if not x or x == '[]':
            return []
        try:
            return json.loads(x)
        except Exception:
            try:
                return eval(x)
            except Exception:
                return []
    return series.apply(_parse)

# Parse span and relation columns
for col in ["affected_spans", "direction_spans", "relations"]:
    silver_df[col] = parse_json_column(silver_df[col])

# Convert date column for temporal split
silver_df['date'] = pd.to_datetime(silver_df['date'])

print(f"Total rows: {len(silver_df):,}")
print(silver_df.head(3)[['header', 'affected_spans', 'direction_spans', 'relations']])
```


## 2. Temporal Split (Train/Val/Test)
Use the same chronological split as the NER notebook: 70% train, 15% val, 15% test without shuffling.


```python

# Sort by date for temporal split
silver_sorted = silver_df.sort_values('date').reset_index(drop=True)

# Optional complexity analysis
silver_sorted['num_dirs'] = silver_sorted['direction_spans'].apply(len)
silver_sorted['num_routes'] = silver_sorted['affected_spans'].apply(len)
silver_sorted['complexity_bin'] = pd.cut(
    silver_sorted['num_dirs'] + silver_sorted['num_routes'],
    bins=[-1, 0, 1, 2, float('inf')],
    labels=['none', 'single', 'double', 'multi']
)

n = len(silver_sorted)
train_end = int(n * 0.70)
val_end = int(n * 0.85)

train_df = silver_sorted.iloc[:train_end].reset_index(drop=True)
val_df = silver_sorted.iloc[train_end:val_end].reset_index(drop=True)
test_df = silver_sorted.iloc[val_end:].reset_index(drop=True)

print(f"Split sizes -> Train: {len(train_df):,}, Val: {len(val_df):,}, Test: {len(test_df):,}")
print("Date ranges:")
print(f"  Train: {train_df['date'].min()} to {train_df['date'].max()}")
print(f"  Val:   {val_df['date'].min()} to {val_df['date'].max()}")
print(f"  Test:  {test_df['date'].min()} to {test_df['date'].max()}")
```


## 3. Tokenization and Vocabularies
Reuse the whitespace/punctuation-aware tokenizer from the NER notebook and build word/character vocabularies from the training split.


```python

from typing import Tuple

# Special tokens
PAD_TOKEN = "<PAD>"
UNK_TOKEN = "<UNK>"
PAD_IDX = 0
UNK_IDX = 1
CHAR_PAD_IDX = 0
CHAR_UNK_IDX = 1

# Relation label map
relation_to_id = {
    "NO_RELATION": 0,
    "HAS_DIRECTION": 1,
}
id_to_relation = {v: k for k, v in relation_to_id.items()}
POSITIVE_LABEL = relation_to_id["HAS_DIRECTION"]


def word_tokenize(text: str) -> List[Tuple[str, int, int]]:
    '''Tokenize text into (token, start_char, end_char) tuples.'''
    tokens = []
    for match in re.finditer(r"\S+", text):
        word = match.group()
        start = match.start()
        for sub in re.finditer(r"[A-Za-z0-9]+|[^\sA-Za-z0-9]", word):
            s = start + sub.start()
            e = s + len(sub.group())
            tokens.append((sub.group(), s, e))
    return tokens

print("Tokenizer sanity check:", word_tokenize("Jamaica-bound J trains are delayed."))

# Build vocabularies from training data
word_counter = Counter()
char_set = set()

for _, row in tqdm(train_df.iterrows(), total=len(train_df), desc="Scanning train text"):
    text = str(row['header']) if pd.notna(row['header']) else ""
    tokens = word_tokenize(text)
    for w, _, _ in tokens:
        wl = w.lower()
        word_counter[wl] += 1
        char_set.update(w)

MIN_WORD_FREQ = 2
word2idx = {PAD_TOKEN: PAD_IDX, UNK_TOKEN: UNK_IDX}
for word, count in word_counter.most_common():
    if count >= MIN_WORD_FREQ:
        word2idx[word] = len(word2idx)

char2idx = {"<PAD>": CHAR_PAD_IDX, "<UNK>": CHAR_UNK_IDX}
for ch in sorted(char_set):
    char2idx[ch] = len(char2idx)

print(f"Word vocab size: {len(word2idx):,}")
print(f"Char vocab size: {len(char2idx):,}")
```


## 4. Dataset Class (with Distance Features)
Create pairwise relation samples, downsample negatives, and compute signed distance between span centers (clipped and embedded later).


```python

MAX_SEQ_LEN = 160
MAX_WORD_LEN = 20
NEGATIVE_RATIO = 1.0
MAX_DIST = 50  # token distance clip for relative position embeddings


def _span_center(mask: List[int]) -> Optional[float]:
    idxs = [i for i, v in enumerate(mask) if v == 1]
    if not idxs:
        return None
    return sum(idxs) / len(idxs)


def _signed_distance(route_mask: List[int], dir_mask: List[int]) -> int:
    r_center = _span_center(route_mask)
    d_center = _span_center(dir_mask)
    if r_center is None or d_center is None:
        return 0
    dist = round(d_center - r_center)
    return max(-MAX_DIST, min(MAX_DIST, dist))


def _distance_id(dist: int) -> int:
    # Map [-MAX_DIST, MAX_DIST] -> [0, 2*MAX_DIST]
    return dist + MAX_DIST


class MTAREDataset(Dataset):
    def __init__(
        self,
        dataframe: pd.DataFrame,
        word2idx: Dict[str, int],
        char2idx: Dict[str, int],
        route_col: str = "affected_spans",
        direction_col: str = "direction_spans",
        relations_col: str = "relations",
        max_seq_length: int = MAX_SEQ_LEN,
        max_word_length: int = MAX_WORD_LEN,
        negative_ratio: float = NEGATIVE_RATIO,
    ):
        self.word2idx = word2idx
        self.char2idx = char2idx
        self.route_col = route_col
        self.direction_col = direction_col
        self.relations_col = relations_col
        self.max_seq_length = max_seq_length
        self.max_word_length = max_word_length
        self.negative_ratio = negative_ratio
        self.samples = self._build_samples(dataframe)

    @staticmethod
    def _normalize_spans(spans: List[Dict]) -> List[Dict]:
        cleaned = []
        for idx, span in enumerate(spans):
            if not isinstance(span, dict) or 'start' not in span or 'end' not in span:
                continue
            span_copy = {**span}
            if 'id' not in span_copy or not isinstance(span_copy['id'], int):
                span_copy['id'] = idx
            cleaned.append(span_copy)
        return cleaned

    @staticmethod
    def _span_mask(tokens: List[Tuple[str, int, int]], span: Dict) -> List[int]:
        mask = [0] * len(tokens)
        start, end = span['start'], span['end']
        for i, (_, s, e) in enumerate(tokens):
            if s >= start and e <= end:
                mask[i] = 1
            elif e > end:
                break
        return mask

    def _build_samples(self, df: pd.DataFrame) -> List[Dict]:
        samples = []
        for row_idx, row in tqdm(df.reset_index(drop=True).iterrows(), total=len(df), desc="Building samples"):
            text = str(row['header']) if pd.notna(row['header']) else ""
            tokens = word_tokenize(text)
            if not tokens:
                continue

            routes = self._normalize_spans(row[self.route_col])
            directions = self._normalize_spans(row[self.direction_col])
            relations = row[self.relations_col] if isinstance(row[self.relations_col], list) else []

            if not routes or not directions:
                continue

            relation_lookup = {
                (rel.get('route_span_id'), rel.get('direction_span_id')): rel.get('type', 'HAS_DIRECTION')
                for rel in relations if isinstance(rel, dict)
            }

            positives = []
            negatives = []

            for r in routes:
                r_mask = self._span_mask(tokens, r)
                if sum(r_mask) == 0:
                    continue
                for d in directions:
                    d_mask = self._span_mask(tokens, d)
                    if sum(d_mask) == 0:
                        continue
                    label_name = relation_lookup.get((r['id'], d['id']), 'NO_RELATION')
                    label_id = relation_to_id.get(label_name, POSITIVE_LABEL)
                    dist = _signed_distance(r_mask, d_mask)
                    sample = {
                        'tokens': tokens,
                        'route_mask': r_mask,
                        'direction_mask': d_mask,
                        'label': label_id,
                        'row_idx': row_idx,
                        'route_id': r['id'],
                        'direction_id': d['id'],
                        'dist_id': _distance_id(dist),
                    }
                    if label_name == 'NO_RELATION':
                        negatives.append(sample)
                    else:
                        positives.append(sample)

            if not positives and not negatives:
                continue

            if positives:
                max_negs = max(int(len(positives) * self.negative_ratio), 1)
            else:
                max_negs = min(3, len(negatives))

            random.shuffle(negatives)
            samples.extend(positives + negatives[:max_negs])
        return samples

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx: int) -> Dict[str, torch.Tensor]:
        sample = self.samples[idx]
        tokens = sample['tokens'][:self.max_seq_length]
        label = sample['label']
        length = len(tokens)

        word_ids = []
        for word, _, _ in tokens:
            wl = word.lower()
            word_ids.append(self.word2idx.get(wl, UNK_IDX))

        char_ids = []
        for word, _, _ in tokens:
            chars = [self.char2idx.get(c, CHAR_UNK_IDX) for c in word[:self.max_word_length]]
            chars += [CHAR_PAD_IDX] * (self.max_word_length - len(chars))
            char_ids.append(chars)

        route_mask = sample['route_mask'][:self.max_seq_length]
        direction_mask = sample['direction_mask'][:self.max_seq_length]

        pad_len = self.max_seq_length - length
        word_ids += [PAD_IDX] * pad_len
        char_ids += [[CHAR_PAD_IDX] * self.max_word_length] * pad_len
        route_mask += [0] * pad_len
        direction_mask += [0] * pad_len

        return {
            'word_ids': torch.tensor(word_ids, dtype=torch.long),
            'char_ids': torch.tensor(char_ids, dtype=torch.long),
            'route_mask': torch.tensor(route_mask, dtype=torch.float),
            'direction_mask': torch.tensor(direction_mask, dtype=torch.float),
            'labels': torch.tensor(label, dtype=torch.long),
            'lengths': torch.tensor(length, dtype=torch.long),
            'sample_ids': torch.tensor(idx, dtype=torch.long),
            'dist_ids': torch.tensor(sample['dist_id'], dtype=torch.long),
        }
```


## 5. Create Datasets and DataLoaders
Instantiate datasets for each split, inspect class balance, and prepare DataLoaders.


```python

BATCH_SIZE = 128

print("Building datasets...")
train_dataset = MTAREDataset(train_df, word2idx, char2idx)
val_dataset = MTAREDataset(val_df, word2idx, char2idx)
test_dataset = MTAREDataset(test_df, word2idx, char2idx)

print(f"Train samples: {len(train_dataset):,}")
print(f"Val samples: {len(val_dataset):,}")
print(f"Test samples: {len(test_dataset):,}")

label_counts = Counter([s['label'] for s in train_dataset.samples])
print("Label distribution (train):")
for label_id, count in label_counts.items():
    print(f"  {id_to_relation[label_id]}: {count:,}")

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)
```


## 6. Model Architecture
BiLSTM encoder with character CNN features. Span representations are computed via masked mean pooling over token states, then paired for relation classification.


```python

class CharCNN(nn.Module):
    def __init__(self, char_vocab_size: int, char_embedding_dim: int = 50, num_filters: int = 50, kernel_sizes: List[int] = [3, 4, 5], dropout: float = 0.3):
        super().__init__()
        self.char_embedding = nn.Embedding(char_vocab_size, char_embedding_dim, padding_idx=CHAR_PAD_IDX)
        self.convs = nn.ModuleList([
            nn.Conv1d(char_embedding_dim, num_filters, ks, padding=ks // 2)
            for ks in kernel_sizes
        ])
        self.dropout = nn.Dropout(dropout)
        self.output_dim = num_filters * len(kernel_sizes)

    def forward(self, char_ids: torch.Tensor) -> torch.Tensor:
        batch_size, seq_len, max_word_len = char_ids.shape
        char_ids = char_ids.view(-1, max_word_len)
        char_emb = self.char_embedding(char_ids)
        char_emb = char_emb.transpose(1, 2)
        conv_outputs = []
        for conv in self.convs:
            conv_out = F.relu(conv(char_emb))
            pooled = F.max_pool1d(conv_out, conv_out.size(2)).squeeze(2)
            conv_outputs.append(pooled)
        char_repr = torch.cat(conv_outputs, dim=1)
        char_repr = self.dropout(char_repr)
        char_repr = char_repr.view(batch_size, seq_len, -1)
        return char_repr


class BiLSTMRelationExtractor(nn.Module):
    def __init__(
        self,
        vocab_size: int,
        char_vocab_size: int,
        num_labels: int,
        word_embedding_dim: int = 128,
        char_embedding_dim: int = 50,
        hidden_dim: int = 256,
        num_layers: int = 2,
        dropout: float = 0.3,
        pretrained_embeddings: Optional[np.ndarray] = None,
        mlp_hidden: int = 256,
        max_dist: int = MAX_DIST,
        dist_emb_dim: int = 32,
    ):
        super().__init__()
        self.word_embedding = nn.Embedding(vocab_size, word_embedding_dim, padding_idx=PAD_IDX)
        if pretrained_embeddings is not None:
            self.word_embedding.weight.data.copy_(torch.from_numpy(pretrained_embeddings))

        self.char_cnn = CharCNN(char_vocab_size, char_embedding_dim, num_filters=50, kernel_sizes=[3, 4, 5], dropout=dropout)
        combined_dim = word_embedding_dim + self.char_cnn.output_dim

        self.lstm = nn.LSTM(
            input_size=combined_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if num_layers > 1 else 0.0,
        )
        self.dropout = nn.Dropout(dropout)

        self.dist_embedding = nn.Embedding(max_dist * 2 + 1, dist_emb_dim)

        pair_dim = hidden_dim * 2
        classifier_input = pair_dim * 4 + dist_emb_dim  # route, direction, abs diff, product, distance emb
        self.classifier = nn.Sequential(
            nn.Linear(classifier_input, mlp_hidden),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(mlp_hidden, num_labels),
        )

    def _encode(self, word_ids: torch.Tensor, char_ids: torch.Tensor) -> torch.Tensor:
        word_emb = self.word_embedding(word_ids)
        char_emb = self.char_cnn(char_ids)
        combined = torch.cat([word_emb, char_emb], dim=-1)
        combined = self.dropout(combined)
        lstm_out, _ = self.lstm(combined)
        lstm_out = self.dropout(lstm_out)
        return lstm_out

    def forward(self, word_ids: torch.Tensor, char_ids: torch.Tensor, route_mask: torch.Tensor, direction_mask: torch.Tensor, dist_ids: torch.Tensor) -> torch.Tensor:
        lstm_out = self._encode(word_ids, char_ids)

        route_mask = route_mask.unsqueeze(-1)
        direction_mask = direction_mask.unsqueeze(-1)

        route_sum = (lstm_out * route_mask).sum(dim=1)
        route_len = route_mask.sum(dim=1).clamp(min=1e-6)
        route_repr = route_sum / route_len

        dir_sum = (lstm_out * direction_mask).sum(dim=1)
        dir_len = direction_mask.sum(dim=1).clamp(min=1e-6)
        direction_repr = dir_sum / dir_len

        dist_emb = self.dist_embedding(dist_ids)

        pair_repr = torch.cat([
            route_repr,
            direction_repr,
            torch.abs(route_repr - direction_repr),
            route_repr * direction_repr,
            dist_emb,
        ], dim=1)

        logits = self.classifier(pair_repr)
        return logits
```


## 7. Training Configuration
Set optimizer, loss with class weights, and helper loops for training and evaluation.


```python

# Initialize embeddings (Xavier uniform) for words
WORD_EMBEDDING_DIM = 128

def initialize_embeddings(vocab_size: int, embedding_dim: int) -> np.ndarray:
    limit = np.sqrt(6.0 / (vocab_size + embedding_dim))
    emb = np.random.uniform(-limit, limit, (vocab_size, embedding_dim)).astype(np.float32)
    emb[PAD_IDX] = 0.0
    return emb

pretrained_embeddings = initialize_embeddings(len(word2idx), WORD_EMBEDDING_DIM)

# Class weights to handle imbalance
label_counts = Counter([s['label'] for s in train_dataset.samples])
weights = []
for i in range(len(relation_to_id)):
    count = label_counts.get(i, 1)
    weight = sum(label_counts.values()) / (len(relation_to_id) * count)
    weights.append(weight)
class_weights = torch.tensor(weights, dtype=torch.float)
print("Class weights:", {id_to_relation[i]: w for i, w in enumerate(class_weights.tolist())})

# Device
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('mps') if torch.backends.mps.is_available() else torch.device('cpu')
print(f"Using device: {device}")

# Model
model = BiLSTMRelationExtractor(
    vocab_size=len(word2idx),
    char_vocab_size=len(char2idx),
    num_labels=len(relation_to_id),
    word_embedding_dim=WORD_EMBEDDING_DIM,
    char_embedding_dim=50,
    hidden_dim=256,
    num_layers=2,
    dropout=0.3,
    pretrained_embeddings=pretrained_embeddings,
    mlp_hidden=256,
    max_dist=MAX_DIST,
    dist_emb_dim=32,
)
model.to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=2)
loss_fct = nn.CrossEntropyLoss(weight=class_weights.to(device))

EPOCHS = 3
PATIENCE = 5
GRAD_CLIP = 5.0

print(f"Model parameters: {sum(p.numel() for p in model.parameters()):,}")
```

```python

from torch.cuda.amp import GradScaler

def train_epoch(model, data_loader, optimizer, scheduler, device, loss_fct, grad_clip=5.0):
    model.train()
    total_loss = 0.0
    scaler = GradScaler()
    for batch in tqdm(data_loader, desc="Training"):
        word_ids = batch['word_ids'].to(device)
        char_ids = batch['char_ids'].to(device)
        route_mask = batch['route_mask'].to(device)
        direction_mask = batch['direction_mask'].to(device)
        labels = batch['labels'].to(device)
        dist_ids = batch['dist_ids'].to(device)

        optimizer.zero_grad()
        with torch.amp.autocast(device_type="cuda", dtype=torch.float16, enabled=torch.cuda.is_available()):
            logits = model(word_ids, char_ids, route_mask, direction_mask, dist_ids)
            loss = loss_fct(logits, labels)

        total_loss += loss.item()
        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        if grad_clip > 0:
            nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
        scaler.step(optimizer)
        scaler.update()
        scheduler.step(loss.item())
    return total_loss / max(len(data_loader), 1)


def evaluate(model, data_loader, device, loss_fct, return_predictions=False):
    model.eval()
    total_loss = 0.0
    preds = []
    labels_all = []
    sample_ids = []
    with torch.no_grad():
        for batch in tqdm(data_loader, desc="Evaluating"):
            word_ids = batch['word_ids'].to(device)
            char_ids = batch['char_ids'].to(device)
            route_mask = batch['route_mask'].to(device)
            direction_mask = batch['direction_mask'].to(device)
            labels = batch['labels'].to(device)
            dist_ids = batch['dist_ids'].to(device)

            with torch.amp.autocast(device_type="cuda", dtype=torch.float16, enabled=torch.cuda.is_available()):
                logits = model(word_ids, char_ids, route_mask, direction_mask, dist_ids)
                loss = loss_fct(logits, labels)
            total_loss += loss.item()

            batch_preds = torch.argmax(logits, dim=1)
            preds.extend(batch_preds.cpu().tolist())
            labels_all.extend(labels.cpu().tolist())
            if return_predictions:
                sample_ids.extend(batch['sample_ids'].cpu().tolist())

    avg_loss = total_loss / max(len(data_loader), 1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels_all, preds, average='binary', pos_label=POSITIVE_LABEL, zero_division=0)
    report = classification_report(labels_all, preds, target_names=[id_to_relation[i] for i in range(len(relation_to_id))])
    results = {
        'loss': avg_loss,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'report': report,
    }
    if return_predictions:
        return results, sample_ids, preds
    return results
```


## 8. Training Loop
Train with early stopping on validation F1.


```python

best_f1 = 0.0
patience_ctr = 0
best_state = None

print("Starting training...")
for epoch in range(EPOCHS):
    print(f"Epoch {epoch + 1}/{EPOCHS}")
    train_loss = train_epoch(model, train_loader, optimizer, scheduler, device, loss_fct, grad_clip=GRAD_CLIP)
    val_results = evaluate(model, val_loader, device, loss_fct)

    print(f"Train loss: {train_loss:.4f}")
    print(f"Val -> loss: {val_results['loss']:.4f}, precision: {val_results['precision']:.4f}, recall: {val_results['recall']:.4f}, f1: {val_results['f1']:.4f}")
    print(val_results['report'])

    if val_results['f1'] > best_f1:
        best_f1 = val_results['f1']
        best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
        patience_ctr = 0
        print(f"New best model (F1={best_f1:.4f})")
    else:
        patience_ctr += 1
        print(f"No improvement. Patience {patience_ctr}/{PATIENCE}")
        if patience_ctr >= PATIENCE:
            print("Early stopping triggered.")
            break

if best_state is not None:
    model.load_state_dict(best_state)
    print(f"Loaded best model with val F1={best_f1:.4f}")
```


## 12. End-to-End Evaluation with BiLSTM NER Spans
Run the RE model using spans predicted by the BiLSTM NER model (instead of gold spans) to approximate true end-to-end performance.


```python

# Load BiLSTM NER model for span prediction
try:
    from torchcrf import CRF
    CRF_AVAILABLE = True
except ImportError:
    CRF_AVAILABLE = False
    CRF = None
    print("pytorch-crf not installed; NER decode will fall back to argmax.")

ner_ckpt_path = "models/bilstm_ner_best/model.pt"
ner_state = torch.load(ner_ckpt_path, map_location='cpu')
ner_word2idx = ner_state['word2idx']
ner_char2idx = ner_state['char2idx']
ner_labels_to_ids = ner_state['labels_to_ids']
ner_ids_to_labels = {int(k): v for k, v in ner_state['ids_to_labels'].items()} if isinstance(ner_state['ids_to_labels'], dict) else ner_state['ids_to_labels']

NER_IGNORE_INDEX = -100
NER_PAD_IDX = ner_word2idx.get('<PAD>', 0)
NER_CHAR_PAD_IDX = ner_char2idx.get('<PAD>', 0)

class NERCharCNN(nn.Module):
    def __init__(self, char_vocab_size: int, char_embedding_dim: int = 50, num_filters: int = 50, kernel_sizes: List[int] = [3, 4, 5], dropout: float = 0.3):
        super().__init__()
        self.char_embedding = nn.Embedding(char_vocab_size, char_embedding_dim, padding_idx=NER_CHAR_PAD_IDX)
        self.convs = nn.ModuleList([
            nn.Conv1d(char_embedding_dim, num_filters, ks, padding=ks // 2)
            for ks in kernel_sizes
        ])
        self.dropout = nn.Dropout(dropout)
        self.output_dim = num_filters * len(kernel_sizes)

    def forward(self, char_ids: torch.Tensor) -> torch.Tensor:
        batch_size, seq_len, max_word_len = char_ids.shape
        char_ids = char_ids.view(-1, max_word_len)
        char_emb = self.char_embedding(char_ids)
        char_emb = char_emb.transpose(1, 2)
        conv_outputs = []
        for conv in self.convs:
            conv_out = F.relu(conv(char_emb))
            pooled = F.max_pool1d(conv_out, conv_out.size(2)).squeeze(2)
            conv_outputs.append(pooled)
        char_repr = torch.cat(conv_outputs, dim=1)
        char_repr = self.dropout(char_repr)
        char_repr = char_repr.view(batch_size, seq_len, -1)
        return char_repr

class BiLSTMCRFNER(nn.Module):
    def __init__(
        self,
        vocab_size: int,
        char_vocab_size: int,
        num_labels: int,
        word_embedding_dim: int = 128,
        char_embedding_dim: int = 50,
        hidden_dim: int = 256,
        num_layers: int = 2,
        dropout: float = 0.3,
        pretrained_embeddings: Optional[np.ndarray] = None,
    ):
        super().__init__()
        self.num_labels = num_labels
        self.word_embedding = nn.Embedding(vocab_size, word_embedding_dim, padding_idx=NER_PAD_IDX)
        if pretrained_embeddings is not None:
            self.word_embedding.weight.data.copy_(torch.from_numpy(pretrained_embeddings))

        self.char_cnn = NERCharCNN(char_vocab_size, char_embedding_dim, num_filters=50, kernel_sizes=[3, 4, 5], dropout=dropout)
        combined_dim = word_embedding_dim + self.char_cnn.output_dim

        self.lstm = nn.LSTM(
            input_size=combined_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if num_layers > 1 else 0.0,
        )
        self.dropout = nn.Dropout(dropout)
        self.hidden2tag = nn.Linear(hidden_dim * 2, num_labels)
        self.crf = CRF(num_labels, batch_first=True) if CRF_AVAILABLE else None

    def _get_emissions(self, word_ids: torch.Tensor, char_ids: torch.Tensor) -> torch.Tensor:
        word_emb = self.word_embedding(word_ids)
        char_emb = self.char_cnn(char_ids)
        combined = torch.cat([word_emb, char_emb], dim=-1)
        combined = self.dropout(combined)
        lstm_out, _ = self.lstm(combined)
        lstm_out = self.dropout(lstm_out)
        emissions = self.hidden2tag(lstm_out)
        return emissions

    def decode(self, word_ids: torch.Tensor, char_ids: torch.Tensor, lengths: torch.Tensor) -> List[List[int]]:
        emissions = self._get_emissions(word_ids, char_ids)
        batch_size, seq_len = word_ids.shape
        mask = torch.arange(seq_len, device=word_ids.device).unsqueeze(0) < lengths.unsqueeze(1)
        if self.crf is not None:
            return self.crf.decode(emissions, mask=mask)
        preds = emissions.argmax(dim=-1).tolist()
        return [p[:l] for p, l in zip(preds, lengths.tolist())]

ner_config = ner_state.get('config', {})
ner_model = BiLSTMCRFNER(
    vocab_size=ner_config.get('vocab_size', len(ner_word2idx)),
    char_vocab_size=ner_config.get('char_vocab_size', len(ner_char2idx)),
    num_labels=len(ner_labels_to_ids),
    word_embedding_dim=ner_config.get('word_embedding_dim', 128),
    char_embedding_dim=ner_config.get('char_embedding_dim', 50),
    hidden_dim=ner_config.get('hidden_dim', 256),
    num_layers=ner_config.get('num_layers', 2),
    dropout=ner_config.get('dropout', 0.3),
)
ner_model.load_state_dict(ner_state['model_state_dict'])
ner_model.to(device)
ner_model.eval()

class NERInferDataset(Dataset):
    def __init__(self, dataframe: pd.DataFrame, word2idx: Dict[str, int], char2idx: Dict[str, int], max_seq_length: int = 128, max_word_length: int = 20):
        self.samples = []
        self.word2idx = word2idx
        self.char2idx = char2idx
        self.max_seq_length = max_seq_length
        self.max_word_length = max_word_length
        for _, row in dataframe.iterrows():
            text = str(row['header']) if pd.notna(row['header']) else ""
            tokens = word_tokenize(text)
            if not tokens:
                continue
            self.samples.append({'tokens': tokens, 'text': text})

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        sample = self.samples[idx]
        tokens = sample['tokens'][:self.max_seq_length]
        length = len(tokens)

        word_ids = []
        char_ids = []
        for word, _, _ in tokens:
            wl = word.lower()
            word_ids.append(ner_word2idx.get(wl, ner_word2idx.get('<UNK>', 1)))
            chars = [ner_char2idx.get(c, ner_char2idx.get('<UNK>', 1)) for c in word[:self.max_word_length]]
            chars += [NER_CHAR_PAD_IDX] * (self.max_word_length - len(chars))
            char_ids.append(chars)

        pad_len = self.max_seq_length - length
        word_ids += [NER_PAD_IDX] * pad_len
        char_ids += [[NER_CHAR_PAD_IDX] * self.max_word_length] * pad_len

        return {
            'word_ids': torch.tensor(word_ids, dtype=torch.long),
            'char_ids': torch.tensor(char_ids, dtype=torch.long),
            'lengths': torch.tensor(length, dtype=torch.long),
            'tokens': tokens,
        }

# Collate to keep token lists variable-length while stacking tensors
def ner_collate_fn(batch):
    word_ids = torch.stack([b['word_ids'] for b in batch])
    char_ids = torch.stack([b['char_ids'] for b in batch])
    lengths = torch.stack([b['lengths'] for b in batch])
    tokens = [b['tokens'] for b in batch]
    return {'word_ids': word_ids, 'char_ids': char_ids, 'lengths': lengths, 'tokens': tokens}

usable_mask = gold_df['affected_spans_gold'].apply(lambda x: isinstance(x, list)) & gold_df['direction_spans_gold'].apply(lambda x: isinstance(x, list))
gold_pred_base = gold_df[usable_mask].reset_index(drop=True)

ner_dataset = NERInferDataset(gold_pred_base, ner_word2idx, ner_char2idx, max_seq_length=128, max_word_length=20)
ner_loader = DataLoader(ner_dataset, batch_size=128, collate_fn=ner_collate_fn)

route_preds = []
direction_preds = []
with torch.no_grad():
    for batch in tqdm(ner_loader, desc="NER decoding"):
        word_ids = batch['word_ids'].to(device)
        char_ids = batch['char_ids'].to(device)
        lengths = batch['lengths'].to(device)
        pred_ids = ner_model.decode(word_ids, char_ids, lengths)
        tokens_batch = batch['tokens']
        for toks, preds_seq in zip(tokens_batch, pred_ids):
            labels_seq = [ner_ids_to_labels[p] for p in preds_seq]
            def _labels_to_spans(tokens, labels, entity_type):
                spans = []
                i = 0
                while i < len(labels):
                    label = labels[i]
                    if label == f'B-{entity_type}':
                        start = tokens[i][1]
                        end = tokens[i][2]
                        value = tokens[i][0]
                        j = i + 1
                        while j < len(labels) and labels[j] == f'I-{entity_type}':
                            end = tokens[j][2]
                            value += ' ' + tokens[j][0]
                            j += 1
                        spans.append({'start': start, 'end': end, 'type': entity_type, 'value': value})
                        i = j
                    else:
                        i += 1
                return spans
            route_preds.append(_labels_to_spans(toks, labels_seq, 'ROUTE'))
            direction_preds.append(_labels_to_spans(toks, labels_seq, 'DIRECTION'))
```

```python

# Build relations from predicted spans (exact match to gold) and evaluate end-to-end
pred_df = gold_pred_base.copy()
pred_df['affected_spans'] = route_preds
pred_df['direction_spans'] = direction_preds

# ensure ids exist
for spans in pred_df['affected_spans']:
    for idx, s in enumerate(spans):
        s['id'] = idx
for spans in pred_df['direction_spans']:
    for idx, s in enumerate(spans):
        s['id'] = idx


def _span_map(spans):
    m = {}
    for s in spans:
        m[(s.get('start'), s.get('end'), s.get('type'))] = s.get('id') if isinstance(s.get('id'), int) else None
    return m

relations_for_pred = []
for _, row in pred_df.iterrows():
    gold_routes = row.get('affected_spans_gold', [])
    gold_dirs = row.get('direction_spans_gold', [])
    gold_rels = row.get('relations_gold', [])
    # parse if strings
    if isinstance(gold_routes, str):
        try:
            gold_routes = json.loads(gold_routes)
        except Exception:
            gold_routes = []
    if isinstance(gold_dirs, str):
        try:
            gold_dirs = json.loads(gold_dirs)
        except Exception:
            gold_dirs = []
    if isinstance(gold_rels, str):
        try:
            gold_rels = json.loads(gold_rels)
        except Exception:
            gold_rels = []

    r_map = _span_map(row['affected_spans'])
    d_map = _span_map(row['direction_spans'])
    mapped = []
    for rel in gold_rels or []:
        gr = rel.get('route_span_id')
        gd = rel.get('direction_span_id')
        r_span = next((s for s in gold_routes if s.get('id') == gr), None)
        d_span = next((s for s in gold_dirs if s.get('id') == gd), None)
        if r_span and d_span:
            r_id = r_map.get((r_span.get('start'), r_span.get('end'), r_span.get('type')))
            d_id = d_map.get((d_span.get('start'), d_span.get('end'), d_span.get('type')))
            if r_id is not None and d_id is not None:
                mapped.append({'route_span_id': r_id, 'direction_span_id': d_id, 'type': 'HAS_DIRECTION'})
    relations_for_pred.append(mapped)

pred_df['relations'] = relations_for_pred

# Filter usable rows
pred_df = pred_df[
    pred_df['affected_spans'].apply(lambda x: isinstance(x, list))
    & pred_df['direction_spans'].apply(lambda x: isinstance(x, list))
].reset_index(drop=True)

end2end_dataset = MTAREDataset(
    pred_df[['header', 'affected_spans', 'direction_spans', 'relations']],
    word2idx,
    char2idx,
    route_col='affected_spans',
    direction_col='direction_spans',
    relations_col='relations',
    max_seq_length=MAX_SEQ_LEN,
    max_word_length=MAX_WORD_LEN,
    negative_ratio=NEGATIVE_RATIO,
)
end2end_loader = DataLoader(end2end_dataset, batch_size=128)

end2end_results, e2e_sample_ids, e2e_preds = evaluate(model, end2end_loader, device, loss_fct, return_predictions=True)
print("END-TO-END RESULTS (BiLSTM NER spans to RE)")
print(f"Loss: {end2end_results['loss']:.4f}")
print(f"Precision: {end2end_results['precision']:.4f}")
print(f"Recall: {end2end_results['recall']:.4f}")
print(f"F1: {end2end_results['f1']:.4f}")
print(end2end_results['report'])

# Relation-level metrics for end-to-end
pred_by_row = {}
for sid, pred in zip(e2e_sample_ids, e2e_preds):
    if pred != POSITIVE_LABEL:
        continue
    sample = end2end_dataset.samples[sid]
    pred_by_row.setdefault(sample['row_idx'], []).append({
        'route_span_id': sample['route_id'],
        'direction_span_id': sample['direction_id'],
        'type': 'HAS_DIRECTION',
    })

def relation_span_key(rel):
    return (rel.get('route_span_id'), rel.get('direction_span_id'), rel.get('type', 'HAS_DIRECTION'))

true_sets = {}
for idx, row in pred_df.reset_index(drop=True).iterrows():
    rels = row['relations'] if isinstance(row['relations'], list) else []
    true_sets[idx] = {relation_span_key(r) for r in rels if isinstance(r, dict)}

pred_sets = {idx: {relation_span_key(r) for r in rels} for idx, rels in pred_by_row.items()}

TP = FP = FN = 0
for idx in true_sets.keys():
    gold_set = true_sets[idx]
    pred_set = pred_sets.get(idx, set())
    TP += len(gold_set & pred_set)
    FP += len(pred_set - gold_set)
    FN += len(gold_set - pred_set)

rel_precision = TP / (TP + FP) if TP + FP > 0 else 0.0
rel_recall = TP / (TP + FN) if TP + FN > 0 else 0.0
rel_f1 = 2 * rel_precision * rel_recall / (rel_precision + rel_recall) if rel_precision + rel_recall > 0 else 0.0

print("Relation-level metrics (end-to-end)")
print(f"Precision: {rel_precision:.4f}, Recall: {rel_recall:.4f}, F1: {rel_f1:.4f} (TP={TP}, FP={FP}, FN={FN})")
```
